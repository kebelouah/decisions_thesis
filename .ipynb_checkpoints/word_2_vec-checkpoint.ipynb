```python
import pandas as pd
from bs4 import BeautifulSoup

from nltk.tokenize import sent_tokenize, word_tokenize
import gensim
from gensim.models import Word2Vec
import logging  # Setting up the loggings to monitor gensim
import re

# Read the HTML file
with open('./final_dataset/570.html', 'r') as file:
    html_content = file.read()
# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')
# Extract the text from the HTML
text = soup.get_text()

list_words = text.replace('\n', ' ').split(' ')
list_words = [word for word in list_words if word != '']
list_words = [word.lower() for word in list_words]
list_words = [word.replace('\xa0', ' ') for word in list_words]
list_words = [re.sub(r'[^\w\s]', '', word) for word in list_words]
list_words = [re.sub(r'\d+', '', word) for word in list_words]
list_words = [word for word in list_words if word != '']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

stop_words = ['απόφασης']

vectorizer = CountVectorizer(stop_words=stop_words)
X = vectorizer.fit_transform(list_words).toarray()
pd.DataFrame(X)
```
